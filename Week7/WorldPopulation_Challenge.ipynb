```python
######## Step 1 ########
'''
import libraries
'''
import requests
from bs4 import BeautifulSoup
```

```python
######## Step 2 ########
'''
Webpage
'''
source_path = "https://www.worldometers.info/world-population/population-by-country/"
page = requests.get(source_path)
page_content = page.content
```

```python
######## Step 3 ########
'''
html.parser
'''
soup = BeautifulSoup(page_content, 'html.parser')
print (soup.prettify())
```

```python
for row in soup.table.find_all('tr'):
#     print (row)
    if row.td == None:
        pass
    else:

#         row_cell = row.td.get_text()
#         print (row_cell)
        for cell in row.find_all('td')[1]:
            print (cell.get_text())
    
```

```python
######## Step 4 ########

#==================================================
# YOUR CODE GOES HERE
#==================================================


# This is how you write to an external tab-delimited file
f = open("world_population.tab", "w", encoding='utf-8')
f.write("rank\tcountry\tpopulation\tyearly_change\tnet_change\tdensity\tland_area\tmigrants\tfert_rate\tmed_age\turban_pop\tworld_share\n")

# What will you write?
for row in soup.table.find_all('tr'):
    if row.td == None:
        pass
    else:
        for i in range(12):
#             print(row.find_all('td')[i])
            if i != 11:
                f.write(row.find_all('td')[i].get_text() + "\t")
            else:
                f.write(row.find_all('td')[i].get_text() + "\n")
    
f.close()
```
